##Bai1:Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó

```python
n = int(input("Nhap chieu dai:"))
m = int(input("Nhap chieu rong: "))
print ("Dien tich hinh chu nhat la : ", n*m )
print ("Chu vi hinh chu nhat la : ", (n+m)*2)
```

##Bai5:Viết chương trình giải phương trình bậc 2 (ax^2 + bx + c = 0)

```python
a,b,c = map(float, input("Nhap 3 so:").split())
delta = b**2 - 4*a*c
if delta < 0: 
    print ("Phuong trinh vo nghiem")
elif delta == 0:
    x = -b/(2*a)
    print ("Phuong trinh co nghiem kep: ", x)
else: 
    x1 = (-b + delta**0.5)/(2*a)
    x2 = (-b - delta**0.5)/(2*a)
    print ("Phuong trinh co 2 nghiem phan biet: ", x1, "va", x2)
```
