##Bai3:Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a,b,c = map(float, input("Nhap 3 canh cua tam giac: ").split())
p = (a+b+c)/2
s = (p*(p-a)*(p-b)*(p-c))**0.5
h = a+b+c
if a+b<=c and b+c<=a and a+c<=b:
    print ("Day khong phai la 3 canh cua tam giac")
elif a==b and b==c:
    print ("Day la tam giac deu.", "Dien tich cua tam giac deu la: ", s , ".Chu vi cua tam giac deu la: ", h)
elif a==b or a==c or b==c:
    print ("Day la tam giac can.", "Dien tich cua tam giac can la: ", s, ".Chu vi cua tam giac can la: ", h)
elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
    print ("Day la tam giac vuong.", "Dien tich cua tam giac vuong la: ", s, ".Chu vi cua tam giac vuong la: ", h)
else:
    print ("Day la tam giac thuong.", "Dien tich cua tam giac thuong la: ", s, ".Chu vi cua tam giac thuong la: ", h)

```

##Bai5:Viết chương trình giải phương trình bậc 2 (ax^2 + bx + c = 0)

```python
a,b,c = map(float, input("Nhap 3 so:").split())
delta = b**2 - 4*a*c
if delta < 0: 
    print ("Phuong trinh vo nghiem")
elif delta == 0:
    x = -b/(2*a)
    print ("Phuong trinh co nghiem kep: ", x)
else: 
    x1 = (-b + delta**0.5)/(2*a)
    x2 = (-b - delta**0.5)/(2*a)
    print ("Phuong trinh co 2 nghiem phan biet: ", x1, "va", x2)
```

##Bai8:Giải hệ phương trình tuyến tính: a*x + b*y = m và
                                        c*x + d*y  = n

```python
a, b, m = map(float, input("Nhap 3 so: ").split())
c, d, n = map(float, input("Nhap 3 so: ").split())
#dùng quy tắc cramer để giải hệ phương trình
#tính các định thức
D = a * d - b * c
Dx = m * d - b* n
Dy = a * n - m * c
if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Heco nghiem duy nhat: x = {x} va y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("He phuong trinh co vo so nghiem")
    else:
        print("He phuong trinh vo nghiem")
```

##Bai10:Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
x0, y0 = map(float, input("Nhap toa do tam I (x0, y0): ").split())
R = float(input("Nhap ban kinh R: "))
x, y = map(float, input("Nhap toa do diem A (x, y): ").split())
AI = math.sqrt((x - x0) ** 2 + (y - y0) ** 2)
if AI < R:
    print(f"Diem A({x}, {y}) nam trong đường tron tam I({x0}, {y0}) và ban kinh R = {R}.")
else:
    print(f"Diem A({x}, {y}) không nam trong đuong tron tam I({x0}, {y0}) và ban kinh R = {R}.")
```
