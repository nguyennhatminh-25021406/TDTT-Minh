##W4A7 "[LargestPrimeFactor]
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 

```python
n = int(input())
uoc_so_nguyen_to_lon_nhat = -1
while n % 2 == 0:
    uoc_so_nguyen_to_lon_nhat = 2
    n //= 2
i = 3
while i * i <= n:
    while n % i == 0:
        uoc_so_nguyen_to_lon_nhat = i
        n //= i
    i += 2
if n > 1:
    uoc_so_nguyen_to_lon_nhat = n

print(uoc_so_nguyen_to_lon_nhat)
```

##W4A8 ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
n = int(input())
steps = 0
while True:
    s = str(n)
    if s == s[::-1]:
        break
    rev = int(s[::-1])
    n = n + rev
    steps += 1
print(steps, n)
```

##W4A9 [PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).


```python
n = int(input())
found = False
for i in range(1, n+1):
    if i*i > n:
        break
    sq = i * i
    s = str(sq)
    ok = True
    for j in range(len(s)):
        for k in range(j+1, len(s)):
            if s[j] == s[k]:
                ok = False
    if ok:
        print(sq, end=" ")
        found = True
if not found:
    print("no number")
```

##W4A10 "[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 


```python
n = int(input())
max_len = 0
result = 1
for i in range(1, n+1):
    x = i
    count = 1
    while x != 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = 3*x + 1
        count += 1
    if count > max_len:
        max_len = count
        result = i
print(result, max_len)
```

##W4A13 [ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a, b = map(int, input().split())
sum_a = 0
for i in range(1, a):
    if a % i == 0:
        sum_a += i
sum_b = 0
for i in range(1, b):
    if b % i == 0:
        sum_b += i
if sum_a == b and sum_b == a:
    print("true")
else:
    print("false")
```

##W4A25 Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
n = int(input())
max_value = n
min_value = n
while n != -1:
    if n > max_value:
        max_value = n
    if n < min_value:
        min_value = n
    n = int(input())
print(f"Giá trị lớn nhất: {max_value}, Giá trị nhỏ nhất: {min_value}")
```

##W4A26 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
x = 1
y = 1
while y <= A:
    temp = y
    y = x + y
    x = temp
print(x)
```

##W4A27 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
