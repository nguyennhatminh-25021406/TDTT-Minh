##W3A3 Kiểm tra xem một số có phải là lũy thừa của 2 hay không Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không. Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n = int(input())
if n & (n - 1) == 0:
    print(f"{n} la luy thua cua 2")
else:
    print(f"{n} khong la luy thua cua 2")
```

##W3A11 [Phân loại tam giác] Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".

```python
a, b, c = map(int, input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print("Tam giac can")
    else:
        print("Tam giac thuong")
else:    
    print("Khong phai tam giac")
```

##W3A13 [Tính tiền điện] Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
0-50 kWh: 1500đ/kWh
51-100 kWh: 2000đ/kWh
100 kWh: 3000đ/kWh

```python
n = int(input())
if n <= 50:
    print("So tien phai tra:", n * 1500)
elif n <= 100:
    print("So tien phai tra:", 50 * 1500 + (n - 50) * 2000)
else:
    print("So tien phai tra:", 50 * 1500 + 50 * 2000 + (n - 100) * 3000)
```
